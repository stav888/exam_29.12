```python
import numpy as np

x = np.array([0, 1, 2, 3])
y = np.array([1, 3, 5, 9])

ŷ = 2 * x + 1

#1 - MSE
mse = np.mean((y - ŷ) ** 2)

#2 - R^2
y_mean = np.mean(y)
sse = np.sum((y - ŷ) ** 2)
sst = np.sum((y - y_mean) ** 2)
r2 = 1 - sse / sst

#3 - Adjusted R^2
n = len(y)
p = 1
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print("ŷ:", ŷ)
print("MSE:", mse)
print("R^2:", r2)
print("Adjusted R^2:", adj_r2)
```
